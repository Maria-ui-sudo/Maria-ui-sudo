Here's a Python function that writes content to a file and then reads it back. This demonstrates basic file handling operations.

###Simple

```python
def simple_file_handler(filename, content_to_write):
    """
    Writes content to a file and then reads it back.

    Args:
        filename (str): The name of the file to create/write to.
        content_to_write (str): The string content to write into the file.

    Returns:
        str: The content read back from the file.
    """
    print(f"Writing to {filename}...")
    try:
        with open(filename, 'w') as f:
            f.write(content_to_write)
        print("Content written successfully.")

        print(f"Reading from {filename}...")
        with open(filename, 'r') as f:
            read_content = f.read()
        print("Content read successfully.")
        return read_content
    except IOError as e:
        return f"An error occurred: {e}"

# Example usage:
file_name = "my_example_file.txt"
text_to_save = "Hello, Colab! This is a test of file handling."

read_data = simple_file_handler(file_name, text_to_save)
print(f"\nContent read from file: '{read_data}'")
```

### Intermediate

This `FileHandler` class provides methods to write text to a specified file and to read text from it. It also includes error handling for `IOError`.

```python
class FileHandler:
    """
    A class to handle basic file operations like writing and reading.
    """

    def __init__(self, filename):
        """
        Initializes the FileHandler with a given filename.

        Args:
            filename (str): The name of the file to operate on.
        """
        self.filename = filename

    def write_content(self, content):
        """
        Writes the given content to the file.

        Args:
            content (str): The string content to write to the file.

        Returns:
            bool: True if write was successful, False otherwise.
        """
        print(f"Writing to {self.filename}...")
        try:
            with open(self.filename, 'w') as f:
                f.write(content)
            print("Content written successfully.")
            return True
        except IOError as e:
            print(f"An error occurred while writing: {e}")
            return False

    def read_content(self):
        """
        Reads all content from the file.

        Returns:
            str: The content read from the file, or an error message if an error occurred.
        """
        print(f"Reading from {self.filename}...")
        try:
            with open(self.filename, 'r') as f:
                read_content = f.read()
            print("Content read successfully.")
            return read_content
        except IOError as e:
            return f"An error occurred while reading: {e}"

# Example usage of the FileHandler class:
my_file = FileHandler("class_example_file.txt")

# Write some content
success = my_file.write_content("This is a test from the FileHandler class.")

if success:
    # Read the content back
    read_data = my_file.read_content()
    print(f"\nContent read from file: '{read_data}'")
```

```python
import os

class FileManagementSystem:
    """
    A class to encapsulate file handling functionalities.
    """
    def __init__(self, base_directory):
        """
        Initializes the FileManagementSystem with a given base directory.
        Ensures the base directory exists, creating it if necessary.

        Args:
            base_directory (str): The path to the base directory where files will be managed.
        """
        self.base_directory = base_directory
        try:
            os.makedirs(self.base_directory, exist_ok=True)
            print(f"Base directory '{self.base_directory}' ensured.")
        except OSError as e:
            print(f"Error creating base directory '{self.base_directory}': {e}")

    def create_or_write_file(self, filename, content):
        """
        Creates a new file or overwrites an existing one with specified content.

        Args:
            filename (str): The name of the file to create or overwrite.
            content (str): The content to write to the file.

        Returns:
            bool: True if the operation was successful, False otherwise.
        """
        file_path = os.path.join(self.base_directory, filename)
        try:
            with open(file_path, 'w') as f:
                f.write(content)
            print(f"Successfully wrote content to '{file_path}'.")
            return True
        except IOError as e:
            print(f"Error writing to file '{file_path}': {e}")
            return False

    def read_file(self, filename):
        """
        Reads and returns the content of a specified file.

        Args:
            filename (str): The name of the file to read.

        Returns:
            str or None: The content of the file if successful, otherwise None.
        """
        file_path = os.path.join(self.base_directory, filename)
        try:
            with open(file_path, 'r') as f:
                content = f.read()
            print(f"Successfully read content from '{file_path}'.")
            return content
        except FileNotFoundError:
            print(f"Error: File '{file_path}' not found.")
            return None
        except IOError as e:
            print(f"Error reading file '{file_path}': {e}")
            return None

    def append_to_file(self, filename, content):
        """
        Appends content to an existing file.

        Args:
            filename (str): The name of the file to append to.
            content (str): The string content to append to the file.

        Returns:
            bool: True if the append operation was successful, False otherwise.
        """
        file_path = os.path.join(self.base_directory, filename)
        try:
            with open(file_path, 'a') as f:
                f.write(content)
            print(f"Successfully appended content to '{file_path}'.")
            return True
        except FileNotFoundError:
            print(f"Error: File '{file_path}' not found. Cannot append.")
            return False
        except IOError as e:
            print(f"Error appending to file '{file_path}': {e}")
            return False

    def delete_file(self, filename):
        """
        Removes a specified file from the managed directory.

        Args:
            filename (str): The name of the file to delete.

        Returns:
            bool: True if the deletion was successful, False otherwise.
        """
        file_path = os.path.join(self.base_directory, filename)
        try:
            os.remove(file_path)
            print(f"Successfully deleted file '{file_path}'.")
            return True
        except FileNotFoundError:
            print(f"Error: File '{file_path}' not found. Cannot delete.")
            return False
        except OSError as e:
            print(f"Error deleting file '{file_path}': {e}")
            return False

    def list_files(self):
        """
        Returns a list of all files present in the managed directory.

        Returns:
            list: A list of filenames in the directory if successful, otherwise an empty list.
        """
        try:
            # Get all entries in the base directory
            entries = os.listdir(self.base_directory)
            # Filter for actual files
            files = [entry for entry in entries if os.path.isfile(os.path.join(self.base_directory, entry))]
            print(f"Successfully listed files in '{self.base_directory}'.")
            return files
        except OSError as e:
            print(f"Error listing files in '{self.base_directory}': {e}")
            return []
```

```python
import requests

def fetch_data_from_api(url):
    """
    Fetches data from a given API URL and returns the JSON response.

    Args:
        url (str): The URL of the API endpoint.

    Returns:
        dict or None: The JSON response as a dictionary if successful, None otherwise.
    """
    print(f"Attempting to fetch data from: {url}")
    try:
        response = requests.get(url)
        response.raise_for_status() # Raise an HTTPError for bad responses (4xx or 5xx)
        print("Data fetched successfully.")
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"An error occurred during the API request: {e}")
        return None

# Example usage:
# Using a public API for testing (e.g., JSONPlaceholder for fake online REST API)
api_url = "https://jsonplaceholder.typicode.com/todos/1"
data = fetch_data_from_api(api_url)

if data:
    print("\nFetched Data:")
    print(data)
else:
    print("Failed to fetch data from the API.")
```
