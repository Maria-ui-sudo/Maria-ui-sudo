#Collecting Data Form a CSV file

```python
import pandas as pd
df=pd.read_csv("sample_data/california_housing_train.csv")
print(df.head())
```

```python
display(df.head(10))
```

# Cleaning Data
##Cleaning date means fixing, correcting,and preparing raw data
### Raw dataos  usually messy:


1. Missing values
2.  Wrong values
3. Duplicates
4. Inconsistent formats




##Solutions:


*   Romove missing rows
*   Fill missing values



```python
df.dropna()
df.fillna(0)
df["housing_median_age"].fillna(df["housing_median_age"].mean())
```

# Removing Duplicates

```python
df["housing_median_age"] = df["housing_median_age"].astype(int)
```

### Fixing Inconsistent Text
#### Text data may not be consistent.
"NY", "New York", "new york"

```python
df["City"]=df["City"].str.lower()
df["City"]=df["City"].str.strip()
```

```python
print(df.columns)
```

##Handling Outliers

Outliers = extreme values.
Age=150
Salary=1000000


Basic Check:

```python
df["housing_median_age"].max()
df["housing_median_age"].min()
```

Remove unrealistic values:

```python
df = df[df["housing_median_age"] <= 100]
```

```python
df_filtered = df[df["housing_median_age"] <= 100]
print("Original DataFrame shape:", df.shape)
print("Filtered DataFrame shape:", df_filtered.shape)
display(df_filtered.head())
```
