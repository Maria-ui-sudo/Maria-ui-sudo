### A Company recrds daily Sale(&euro;) for week:
|Day|Sales|(&euro;)
|----|-------------|
|Monday| 200|
|Tuesday| 250|
|Wednesday| 180 |
|Thurday|300|
|Friday|270|


@ Task:
1. Calculate the "total sales" for week
2. Calculate the "average(mean)daily sales"
3. Identify the "highest" and "lowest" sales day
4. Explain what the aveage sales number tells the business

```python

import pandas as pd
import numpy as np
sale= [200,250,180,300,270]
sale_series=pd.Series(sale)
print(sale_series)

```

```python
import pandas as pd
import numpy as np
np.average(np.mean(sale_series))

```

```python
import pandas as pd
import numpy as np

# Using .max() and .min() to find the highest and lowest values
highest_sale = sale_series.max()
lowest_sale = sale_series.min()

print(f"Highest Sale: {highest_sale}")
print(f"Lowest Sale: {lowest_sale}")
```

```python
# Calculate Total Sales
total_sales = sale_series.sum()
print(f"Total Sales for the week: €{total_sales}")
```

### Exercise 3: Final Results

1. **Total Sales for the week:** €1,200
2. **Average (Mean) Daily Sales:** €240
3. **Highest Sales Day:** Thursday (€300)
4. **Lowest Sales Day:** Wednesday (€180)

#### 4. Business Interpretation of Average Sales
The average sales figure (€240) represents the **typical performance** of the shop per day. It allows the business to:
* **Benchmark Performance:** See which days are performing above or below 'normal'.
* **Financial Planning:** Predict future revenue (e.g., expecting ~€1,200 for a 5-day week).
* **Staffing/Inventory:** Helps determine how much stock is needed on a typical day.
