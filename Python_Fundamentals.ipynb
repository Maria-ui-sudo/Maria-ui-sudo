```python
import matplotlib.pyplot as plt
import pandas as pd

sales = [120, 200, 150, 300, 180] # Defined sales list based on the text cell and added 180

sales_series= pd.Series(sales)
plt.figure(figsize=(10,6))
sales_series.plot(kind='bar', color='skyblue')
plt.title('Daily Sales Data')
plt.xlabel('Day Index')
plt.ylabel('Sales Value')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Addressing the remaining requests:
print(f"The first sales value is: {sales[0]}")
print(f"There are {len(sales)} sales values in the list.")
```
